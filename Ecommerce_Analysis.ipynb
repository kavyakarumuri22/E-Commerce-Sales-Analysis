```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display graphs inside the notebook
%matplotlib inline

print("Libraries imported successfully!")
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
translation = pd.read_csv("product_category_name_translation.csv")

print("All datasets loaded successfully!")
```

```python
# Display first 5 rows of each dataset

print("Orders Dataset")
display(orders.head())

print("Order Items Dataset")
display(order_items.head())

print("Products Dataset")
display(products.head())

print("Customers Dataset")
display(customers.head())

print("Reviews Dataset")
display(reviews.head())
```

```python
# Check information about each dataset

print("Orders")
orders.info()

print("\nOrder Items")
order_items.info()

print("\nProducts")
products.info()

print("\nCustomers")
customers.info()

print("\nReviews")
reviews.info()
```

```python
# Check missing values in each dataset

print("Orders Missing Values")
print(orders.isnull().sum())

print("\nOrder Items Missing Values")
print(order_items.isnull().sum())

print("\nProducts Missing Values")
print(products.isnull().sum())

print("\nCustomers Missing Values")
print(customers.isnull().sum())

print("\nReviews Missing Values")
print(reviews.isnull().sum())
```

```python
# Remove duplicate rows

orders = orders.drop_duplicates()

order_items = order_items.drop_duplicates()

products = products.drop_duplicates()

customers = customers.drop_duplicates()

reviews = reviews.drop_duplicates()

print("Duplicates removed successfully!")
```

```python
# Convert order dates

orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])

orders["order_approved_at"] = pd.to_datetime(orders["order_approved_at"])

orders["order_delivered_carrier_date"] = pd.to_datetime(orders["order_delivered_carrier_date"])

orders["order_delivered_customer_date"] = pd.to_datetime(orders["order_delivered_customer_date"])

orders["order_estimated_delivery_date"] = pd.to_datetime(orders["order_estimated_delivery_date"])

print("Date columns converted successfully!")
```

```python
# Fill missing values in products dataset

products["product_category_name"] = products["product_category_name"].fillna("Unknown")

products = products.fillna(0)

# Fill missing review comments

reviews["review_comment_title"] = reviews["review_comment_title"].fillna("No Title")

reviews["review_comment_message"] = reviews["review_comment_message"].fillna("No Comment")

print("Missing values handled successfully!")
```

```python
# Merge orders with order items
df = pd.merge(orders, order_items, on="order_id", how="inner")

# Merge with products
df = pd.merge(df, products, on="product_id", how="left")

# Merge with customers
df = pd.merge(df, customers, on="customer_id", how="left")

# Merge with reviews
df = pd.merge(df, reviews, on="order_id", how="left")

# Merge with category translation
df = pd.merge(df, translation, on="product_category_name", how="left")

print("Datasets merged successfully!")
```

```python
# Display first 5 rows
df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df["Revenue"] = df["price"] + df["freight_value"]

print("Revenue column created successfully!")
```

```python
df["Month"] = df["order_purchase_timestamp"].dt.month_name()

print("Month column created successfully!")
```

```python
# Revenue by Product Category

category_revenue = df.groupby("product_category_name_english")["Revenue"].sum().sort_values(ascending=False)

category_revenue.head(10)
```

```python
plt.figure(figsize=(12,6))

category_revenue.head(10).plot(kind="bar")

plt.title("Top 10 Product Categories by Revenue")

plt.xlabel("Product Category")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

```python
monthly_sales = df.groupby("Month")["Revenue"].sum()

monthly_sales
```

```python
plt.figure(figsize=(12,6))

monthly_sales.plot(marker="o")

plt.title("Monthly Revenue")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.grid(True)

plt.show()
```

```python
state_sales = df.groupby("customer_state")["Revenue"].sum().sort_values(ascending=False)

state_sales.head(10)
```

```python
plt.figure(figsize=(12,6))

state_sales.head(10).plot(kind="bar", color="green")

plt.title("Top 10 States by Revenue")

plt.xlabel("State")

plt.ylabel("Revenue")

plt.xticks(rotation=0)

plt.show()
```

```python
average_order_value = df.groupby("Month")["Revenue"].mean()

average_order_value
```

```python
plt.figure(figsize=(12,6))

average_order_value.plot(marker="o", color="red")

plt.title("Average Order Value Trend")

plt.xlabel("Month")

plt.ylabel("Average Order Value")

plt.grid(True)

plt.show()
```

```python
review_distribution = df["review_score"].value_counts().sort_index()

review_distribution
```

```python
plt.figure(figsize=(7,7))

review_distribution.plot(kind="pie", autopct="%1.1f%%")

plt.title("Customer Review Score Distribution")

plt.ylabel("")

plt.show()
```

```python
pivot_table = pd.pivot_table(
    df,
    values="Revenue",
    index="product_category_name_english",
    columns="Month",
    aggfunc="sum"
)

plt.figure(figsize=(14,16))

sns.heatmap(pivot_table, cmap="YlGnBu")

plt.title("Category vs Month Revenue Heatmap")

plt.show()
```

```python
# KPI Values

total_revenue = df["Revenue"].sum()
total_orders = df["order_id"].nunique()
top_category = category_revenue.idxmax()
best_month = monthly_sales.idxmax()
average_review = df["review_score"].mean()

print("========== E-COMMERCE SALES DASHBOARD ==========")
print(f"Total Revenue      : {total_revenue:,.2f}")
print(f"Total Orders       : {total_orders}")
print(f"Top Category       : {top_category}")
print(f"Best Month         : {best_month}")
print(f"Average Review     : {average_review:.2f}")
```

```python
plt.figure(figsize=(16,10))

plt.suptitle("E-Commerce Sales Dashboard", fontsize=20)

# Top Categories
plt.subplot(2,2,1)
category_revenue.head(5).plot(kind="bar")
plt.title("Top 5 Categories")
plt.xlabel("")
plt.ylabel("Revenue")

# Monthly Revenue
plt.subplot(2,2,2)
monthly_sales.plot(marker="o")
plt.title("Monthly Revenue")
plt.xlabel("")
plt.ylabel("Revenue")

# Top States
plt.subplot(2,2,3)
state_sales.head(5).plot(kind="bar")
plt.title("Top 5 States")
plt.xlabel("")
plt.ylabel("Revenue")

# Review Scores
plt.subplot(2,2,4)
review_distribution.plot(kind="pie", autopct="%1.1f%%")
plt.title("Review Scores")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

# Business Insights Report

## 1. Highest Revenue Category
The highest revenue was generated by the top product category. This indicates strong customer demand for this category. The company should increase inventory levels and invest more in marketing this category to maximize future sales.

## 2. Peak Sales Month
The highest sales occurred during the best-performing month. The company should plan promotional campaigns and ensure sufficient stock before this month to capitalize on increased customer demand.

## 3. Best Performing Region
The top-performing state generated the highest revenue. Expanding logistics, improving delivery speed, and launching targeted advertisements in this region could further increase sales.

## 4. Customer Review Analysis
Most customers gave ratings of 4 and 5 stars, showing high customer satisfaction. Maintaining product quality and customer service will help sustain positive reviews and customer loyalty.

## 5. Average Order Value
The average order value remained relatively stable throughout the year. Introducing product bundles, discounts, and cross-selling strategies can encourage customers to spend more per order.

# Most Surprising Finding

The most surprising finding from this analysis was that a small number of product categories contributed a significant portion of the overall revenue. Additionally, most customers provided positive review scores, indicating strong customer satisfaction. This suggests that focusing on high-performing product categories while maintaining excellent customer service can significantly improve business growth.
